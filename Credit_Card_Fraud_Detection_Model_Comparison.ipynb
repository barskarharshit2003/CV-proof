```python
import numpy as np 
import pandas as pd 
df=pd.read_csv('creditcard.csv')
pd.options.display.max_columns=None
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler()
df['Amount']=sc.fit_transform(pd.DataFrame(df['Amount'])) 
df.head()
```

```python
df.drop('Time',axis=1,inplace=True)
```

```python
df.duplicated().any()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().any()
```

```python
df.shape
```

```python
x=df.drop('Class',axis=1)
y=df['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score 
```

```python
classifier ={ 
        "Logistic Regression" : LogisticRegression() ,
        "Decision Tree Classifier" : DecisionTreeClassifier() ,        
        "Naive Bayes" : GaussianNB()
    }
```

```python
for name,mod in classifier.items(): 
    print(f"____________ {name}______________")
    mod.fit(x_train,y_train)
    y_pred=mod.predict(x_test)
    print(f"Accuracy Score : {accuracy_score(y_test,y_pred)}")
    print(f"Precision Score : {precision_score(y_test,y_pred)}")
    print(f"Recall Score : {recall_score(y_test,y_pred)}")
    print(f"F1 Score : {f1_score(y_test,y_pred)}")
```

```python
#Checking if the data is imbalanced 
df['Class'].value_counts()
#Yes its an imbalanced dataset
```

```python
valid=df[df['Class']==0]
fraud=df[df['Class']==1]
print(f"{valid.shape}")
print(f"{fraud.shape}")
```

```python
valid_undersampling = valid.sample(len(fraud))
print(valid_undersampling.shape)
```

```python
df1=pd.concat([valid_undersampling,fraud],ignore_index=True)
df1.shape
```

```python
x=df1.drop('Class',axis=1)
y=df1['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score 
```

```python
classifier ={ 
        "Logistic Regression" : LogisticRegression() ,
        "Decision Tree Classifier" : DecisionTreeClassifier() , 
        "Naive Bayes" : GaussianNB()    
    }
```

```python
for name,mod in classifier.items(): 
    print(f"____________ {name}______________")
    mod.fit(x_train,y_train)
    y_pred=mod.predict(x_test)
    print(f"Accuracy Score : {accuracy_score(y_test,y_pred)}")
    print(f"Precision Score : {precision_score(y_test,y_pred)}")
    print(f"Recall Score : {recall_score(y_test,y_pred)}")
    print(f"F1 Score : {f1_score(y_test,y_pred)}")
```
